```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the dataframe
df = pd.read_csv("C:\\Users\\Padraig\\Desktop\\Car_Project\\Preprocessing\\Filtered.csv")

# Check the shape of the dataframe
print(f"Shape of dataframe: {df.shape}")
```

```python
# Check the first 5 rows of the dataframe
print(df.head())

# Check the data types of each column
print(df.dtypes)

# Check for missing values in the dataframe
print(df.isnull().sum())

# Check the summary statistics for numerical columns
print(df.describe())
```

```python
# Check the unique values for categorical columns
print(df['Make'].unique())
print(df['Model'].unique())
print(df['Fuel Type'].unique())
print(df['Current Country of Reg.'].unique())
print(df['Imported'].unique())
# Check the correlation between numerical columns
print(df.corr())
```

```python
plt.hist(df['Mileage'], bins=range(0, 300, 10))
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.title('Histogram of Mileage')
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Year'], bins=range(1900, 2025, 5))
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Histogram of Year')
plt.grid(True)
plt.show()

plt.hist(df['Engine Size (Litres)'], bins=np.arange(0.5, 5, 0.1))
plt.xlabel('Engine Size (Litres)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.hist(df['Price'], bins=range(0, 250000, 5000))
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
plt.grid(True)
plt.show()
```

```python
plt.boxplot(df.groupby('Fuel Type')['Price'].apply(list))
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('Box Plot of Price by Fuel Type')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Mileage'], df['Price'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Scatter Plot of Price by Mileage')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Total Owners'].unique(), df['Total Owners'].value_counts())
plt.xlabel('Total Owners')
plt.ylabel('Frequency')
plt.title('Bar Plot of Total Owners')
plt.grid(True)
plt.show()
```

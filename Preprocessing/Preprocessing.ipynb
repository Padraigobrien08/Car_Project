```python
def extract_make_with_index(df, column_name):
    makes = []
    make_indexes = []
    for index, row in df.iterrows():
        found = False
        for i, value in enumerate(row):
            if pd.notnull(value) and value == column_name:
                found = True
                make = row[i+1] if i+1 < len(row) else None
                makes.append(make)
                make_indexes.append(i+1)
                break
        if not found:
            makes.append(None)
            make_indexes.append(None)
    df[column_name] = makes

# Example usage:
# Assuming df is your DataFrame
# Extracting 'Make' column and its index
for col in full_columns:
    print(col)
    extract_make_with_index(df, col)

df
```

```python
def extract_prices(df):
    prices = []
    for index, row in df.iterrows():
        price_found = False
        for value in row:
            if isinstance(value, str) and ('€' in value or '£' in value):
                prices.append(value)
                price_found = True
                break
        if not price_found:
            prices.append(None)
    return prices

df['Price'] = extract_prices(df_copy)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

def median_impute(df, columns):
    new_df = pd.DataFrame(columns=df.columns)
    median_imputer = SimpleImputer(strategy='median')
    median_imputer.fit(df[columns])
    imputed_values = median_imputer.transform(df[columns])
    imputed_values = np.hstack((imputed_values))
    new_df[columns] = imputed_values.reshape(-1, len(columns))
    return new_df

columns_to_impute = ['Total Owners', 'Engine Size (Litres)']
imputed_df = median_impute(df, columns_to_impute)
imputed_df
```

```python
df['Total Owners'] = imputed_df['Total Owners']
df['Engine Size (Litres)'] = imputed_df['Engine Size (Litres)']
```

```python
#'€' in value or '£'
price = []
for i in df["Price"]:
    try:
        i = i.replace(",", "")
        if i[:1] == '€':
            price.append(int(i[1:]))
        elif i[:1] == '£':
            number = int(i[1:])
            number = round(number * 1.17)
            price.append(number)
    except TypeError:
        price.append(None)

df['Price'] = price
price
```

```python
kms = []
for i in df["Mileage"]:
    try:
        i = i.replace(",", "")
        # print(int(i[:-3]), i, i[-2:])
        if i[-2:] == 'km':
            kms.append(int(i[:-3]))
        elif i[-2:] == 'mi':
            number = int(i[:-3])
            number = number * 1.609
            number = round(number)
            
            kms.append(number)
    except TypeError:
        kms.append(None)

df['Mileage'] = kms
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv("C:\\Users\\Padraig\\Desktop\\Car_Project\\Preprocessing\\Filtered.csv")
df
```

```python
# Filter out rows where the price is under 2000
df = df[df['Price'] >= 2000]
```

```python
df[df['Make'] == "Rover"]
```

```python
#print(df['Make'].unique())

# Group by 'Make' and calculate the average cost for each make
average_cost_per_make = df.groupby('Make')['Price'].mean()

# Print out the result
#print(average_cost_per_make)
ttf_df = df[df['Year'].isin([2023, 2024])]
New_Price_Per_Make = ttf_df.groupby('Make')['Price'].mean()
total_price_per_year_make = df.groupby('Make')['Price'].mean()
total_price_per_year_make = New_Price_Per_Make - total_price_per_year_make
# Calculate the total number of cars produced for each year and make
total_cars_per_year_make = abs(2024 - df.groupby('Make')['Year'].mean())

# Calculate the average price per year for each make
average_price_per_year_make = round(total_price_per_year_make / total_cars_per_year_make)
# print(average_price_per_year_make)
# Normalize the average price per year using MinMaxScaler
print(type(average_price_per_year_make))
average_price_per_year_make.replace([np.inf, -np.inf], 50000, inplace=True)
scaler = MinMaxScaler(feature_range=(0, 10))
normalized_average_price_per_year_make = scaler.fit_transform(average_price_per_year_make.values.reshape(-1, 1))
normalized_average_price_per_year_make = np.nan_to_num(normalized_average_price_per_year_make, nan=0)
```

```python
Luxury_encoding = {}
for make, luxury_value in zip(Car_Brands, normalized_average_price_per_year_make):
    Luxury_encoding[make] = round(float(luxury_value), 2)

Luxury_encoding
```

```python
df['Depreciation'] = df['Make'].map(Luxury_encoding)
```
